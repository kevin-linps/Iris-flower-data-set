# Iris Flower Data Set 01 - Visualization

The Iris Flower Data Set is a famous multivariate data set in statistics. It describes three species of Iris flowers using four characteristics â€” sepal length, sepal width, petal length and petal width. Since there are four variables, this data set is four-dimensional. We cannot plot all variables at once, but we can plot multiple graphs can observe the trends.

More information about Iris Flower Data Set: https://en.wikipedia.org/wiki/Iris_flower_data_set

## Plot flower traits using scatter plots

Conveniently scikit-learn package in Python already contains the data set, so we don't need to download it elsewhere. Let's start by including the libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets

iris = datasets.load_iris()
```

For our first graph, let's plot sepal length vs sepal width.

```python
x = [ iris.data[:, :2], iris.data[:, 2:] ] 
y = iris.target
name_x = (iris.feature_names[0], iris.feature_names[2])
name_y = (iris.feature_names[1], iris.feature_names[3])
n_classes = len(iris.target_names)

# Set-up 1x2 grid for plotting.
fig, sub = plt.subplots(1, 2)
plt.subplots_adjust(wspace=0.8, hspace=0.4)

for x_val, nx, ny, ax in zip(x, name_x, name_y, sub.flatten()):
    
    for i, colour in zip(range(n_classes), plt.cm.Set1.colors):
        idx = np.where(y == i)
        ax.scatter(x_val[idx, 0], x_val[idx, 1], c=[colour], 
                   label=iris.target_names[i], edgecolors = 'k', s = 40)
    ax.set_xlabel(nx)
    ax.set_ylabel(ny)

fig.suptitle("Feature Comparison of 3 Iris Species")
plt.legend(title = "Iris Species")
plt.show()
```

We can peek on the answer to see if we graph it right? Yes, we did.

## Plot flower traits in 3D space

The 2D scatter plots above each shows the relationship between two flower traits. Thus, if we were to plot in 3D, then we can display the relationship between three traits at once. Let's explore this!

```python
# set up the figure
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)

X = iris.data[:, :3]    # grab the first three traits

# plot the data points
# ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.Set1, edgecolor='k', s=40)
for i, colour in zip(range(n_classes), plt.cm.Set1.colors):
    idx = np.where(y == i)
    ax.scatter(X[idx, 0], X[idx, 1], X[idx, 2], c=[colour], 
               label=iris.target_names[i], edgecolors = 'k', s = 40)

# axis titles
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_zlabel(iris.feature_names[2])

fig.suptitle("Feature Comparison of 3 Iris Species")
plt.legend(title = "Iris Species")
plt.show()
```

From this graph, we can easily see that the red iris differs from other irises by much shorter sepal length. Although we can see this distinction from the petal length vs petal width graph as well, plotting in 3D space does give us a better understanding about the data set.
